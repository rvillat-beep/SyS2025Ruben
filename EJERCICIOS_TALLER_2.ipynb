# Análisis Comparativo entre las Representaciones de Fourier y el Algoritmo de Transformada Rápida (FFT)

Las distintas formulaciones de Fourier constituyen herramientas fundamentales para el estudio de señales en el dominio de la frecuencia. A continuación, se presentan las semejanzas y diferencias entre la Serie de Fourier, la Transformada de Fourier, la Transformada de Fourier en Tiempo Discreto (DTFT) y la Transformada Discreta de Fourier (DFT), considerando el tipo de señal (tiempo continuo o discreto) y la naturaleza de su espectro (continuo o discreto). También se analiza el funcionamiento y la relevancia computacional del algoritmo de Transformada Rápida de Fourier (FFT), empleado para optimizar el cálculo de la DFT.

## Serie de Fourier (Exponencial, Trigonométrica y Compacta)

**Tipo de señal:**  
Se aplica exclusivamente a señales periódicas definidas en el dominio del tiempo continuo.

**Estructura espectral:**  
El análisis genera un espectro discreto, compuesto por armónicos que son múltiplos enteros de la frecuencia fundamental.

**Semejanzas:**  
Las tres representaciones —exponencial, trigonométrica y compacta— son equivalentes matemáticamente. Cada una describe la misma señal, solo que en diferentes formas. La expresión exponencial, basada en funciones complejas, es la más simple desde el punto de vista analítico; las demás se derivan de ella.

**Diferencias:**  
- La forma trigonométrica expresa la señal mediante senos y cosenos.  
- La forma exponencial utiliza exponentes complejos del tipo e^{jωt}.  
- La forma compacta combina información de amplitud y fase, facilitando la interpretación física de la señal.

## Transformada de Fourier (TF)

**Tipo de señal:**  
Se emplea en señales no periódicas del dominio continuo.

**Estructura espectral:**  
Produce un espectro continuo, que abarca todas las frecuencias reales posibles.

**Semejanzas:**  
Puede entenderse como una extensión de la Serie de Fourier. Cuando el período de una señal periódica tiende al infinito, la Serie de Fourier se transforma en la TF.

**Diferencias:**  
La Serie de Fourier se aplica a señales periódicas y genera espectros discretos; la Transformada de Fourier se aplica a señales no periódicas y produce espectros continuos.

## Transformada de Fourier en Tiempo Discreto (DTFT)

**Tipo de señal:**  
Está diseñada para señales no periódicas en tiempo discreto.

**Estructura espectral:**  
El resultado es un espectro continuo pero periódico en la frecuencia, que se repite cada 2π radianes.

**Semejanzas:**  
Es el equivalente discreto de la Transformada de Fourier, adaptada al análisis de secuencias digitales.

**Diferencias:**  
Aunque el espectro de la DTFT es continuo como el de la TF, se distingue por su periodicidad. Además, la DTFT se aplica a secuencias discretas (teóricamente infinitas), mientras que la TF opera sobre funciones continuas.

## Transformada Discreta de Fourier (DFT)

**Tipo de señal:**  
Se aplica a secuencias discretas y finitas de tiempo. En la práctica, se asume que la secuencia es periódica.

**Estructura espectral:**  
El espectro obtenido es totalmente discreto, compuesto por un número finito de muestras en frecuencia.

**Semejanzas:**  
La DFT puede interpretarse como una versión muestreada de la DTFT, calculada en frecuencias equiespaciadas. Representa una herramienta esencial en el procesamiento digital de señales por su implementación computacional práctica.

**Diferencias:**  
A diferencia de la DTFT, que produce un espectro continuo y periódico, la DFT trabaja sobre secuencias finitas y genera un conjunto discreto de frecuencias.

## Algoritmo de Transformada Rápida de Fourier (FFT)

**Concepto y propósito:**  
La FFT no constituye una nueva transformada, sino un algoritmo eficiente diseñado para calcular la DFT y su inversa de forma rápida. Su objetivo es reducir de manera significativa la cantidad de operaciones necesarias para obtener el espectro discreto de una señal.

**Importancia:**  
Gracias a su eficiencia, la FFT ha permitido el desarrollo de múltiples aplicaciones en ingeniería y tecnología, tales como la compresión digital de audio e imágenes, el análisis espectral en sistemas de comunicación, el procesamiento biomédico y el diseño de filtros digitales.

## Funcionamiento del Algoritmo FFT

El algoritmo FFT más utilizado es el de Cooley-Tukey, basado en el principio de 'divide y vencerás'.  
La idea principal consiste en descomponer una DFT de tamaño N (preferiblemente una potencia de 2) en varias DFT más pequeñas, normalmente de tamaño N/2.  
Esta división aprovecha las simetrías y periodicidades de los factores exponenciales complejos (raíces de la unidad) presentes en la definición de la DFT.  

El proceso continúa recursivamente hasta alcanzar DFTs de tamaño 1, que se calculan de manera trivial. Finalmente, los resultados de las subtransformadas se combinan para reconstruir la DFT completa.  
Existen distintas variantes del algoritmo, como las versiones radix-2 y radix-4, optimizadas según el tamaño de la señal y la arquitectura del sistema computacional.

## Costo Computacional: DFT vs FFT

**DFT directa:**  
El cálculo directo de la DFT requiere un número de operaciones proporcional a O(N²), lo que se vuelve ineficiente para secuencias grandes.

**FFT:**  
El algoritmo FFT reduce la complejidad a O(N log₂ N), ofreciendo una mejora drástica en la velocidad de procesamiento.

**Ejemplo:**  
Para una secuencia de N = 1024 puntos:  
- La DFT directa realiza aproximadamente 1,048,576 operaciones.  
- La FFT ejecuta el mismo cálculo en solo unas 10,240 operaciones.  

Esta enorme diferencia justifica el uso casi universal de la FFT en el procesamiento digital moderno, donde el rendimiento y la eficiencia son factores esenciales.

\begin{cases}

\end{cases}
$$

Transformada:

Resolviendo:

# Modulación por Amplitud con Detección Coherente (AM)

La **modulación por amplitud (AM)** consiste en variar la **amplitud de una portadora senoidal** en función de la señal de mensaje (o información).

La señal modulada se expresa como:

$$
s(t) = [1 + m(t)] \cos(2 \pi f_c t)
$$

donde:

- \( m(t) \): señal mensaje (normalizada entre -1 y 1)
- \( f_c \): frecuencia de la portadora
- \( s(t) \): señal modulada en amplitud

El **índice de modulación** se define como:

$$
\mu = \frac{A_m}{A_c}
$$

donde \( A_m \) es la amplitud de la señal mensaje y \( A_c \) la de la portadora. Este índice mide la relación entre ambas amplitudes.

---

## Detección coherente

Para recuperar la señal original, se multiplica la señal AM por una portadora local sincronizada en **frecuencia y fase**:

$$
r(t) = s(t) \cdot \cos(2 \pi f_c t)
$$

Luego, se pasa por un **filtro pasa-bajas** para eliminar las componentes de alta frecuencia, obteniendo nuevamente la señal del mensaje \( m(t) \).

---

## Aplicaciones

- Radiodifusión AM (radio comercial)
- Comunicación aérea
- Sistemas de transmisión de voz analógica
- Instrumentación y telemetría analógica

```
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de simulación
Fs = 5000           # Frecuencia de muestreo [Hz]
t = np.linspace(0, 0.02, int(Fs*0.02), endpoint=False)  # 20 ms de duración
fc = 500            # Frecuencia de portadora [Hz]
fm = 50             # Frecuencia de mensaje (para el coseno) [Hz]

# --- Entrada del usuario ---
mu = float(input("Ingrese el índice de modulación (μ, recomendado entre 0 y 1): "))

# Amplitudes
Am = 1
Ac = 1

# Señal mensaje tipo pulso rectangular
m_rect = Am * ((t > 0.005) & (t < 0.015)).astype(float)

# Señal mensaje tipo coseno
m_cos = Am * np.cos(2 * np.pi * fm * t)

# Señales moduladas AM
s_rect = Ac * (1 + mu * m_rect) * np.cos(2 * np.pi * fc * t)
s_cos  = Ac * (1 + mu * m_cos)  * np.cos(2 * np.pi * fc * t)

# Función para graficar espectro con rFFT
def plot_spectrum(signal, Fs, title):
    N = len(signal)
    f = np.fft.rfftfreq(N, 1/Fs)
    S = np.abs(np.fft.rfft(signal)) / N
    plt.plot(f, S)
    plt.title(title)
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('|S(f)|')
    plt.grid(True)

# ============================================================
# GRAFICACIÓN DE SEÑALES EN EL TIEMPO
# ============================================================
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t * 1000, m_rect)
plt.title('Mensaje tipo pulso rectangular')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t * 1000, s_rect)
plt.title(f'Señal AM (pulso rectangular, μ={mu})')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t * 1000, m_cos)
plt.title('Mensaje tipo coseno')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t * 1000, s_cos)
plt.title(f'Señal AM (coseno, μ={mu})')
plt.xlabel('Tiempo [ms]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# GRAFICACIÓN DE ESPECTROS EN FRECUENCIA (usando rFFT)
# ============================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_spectrum(s_rect, Fs, 'Espectro de la señal AM (pulso rectangular)')

plt.subplot(1, 2, 2)
plot_spectrum(s_cos, Fs, 'Espectro de la señal AM (coseno)')

plt.tight_layout()
plt.show()
```

```
# Cuaderno Colab: Simulación AM (descarga fragmento de YouTube 20-25s, modulación índice=1, demodulación)
# Requisitos: Colab (tiene ffmpeg). Se instalará yt-dlp para descargar audio.
# Pegar todo en una celda y ejecutar.

# ---------------------------
# 1) Instalación y descargas
# ---------------------------
!pip install -q yt-dlp
!pip install -q soundfile
!pip install -q pydub
# ffmpeg ya suele estar instalado en Colab; si no, descomentar:
# !apt-get update -qq && apt-get install -y -qq ffmpeg

# ---------------------------
# 2) Parámetros del usuario
# ---------------------------
# Pega aquí el link de YouTube que quieres usar:
YT_URL = "https://www.youtube.com/watch?v=Dwzk-XZxZ4k&list=RDDwzk-XZxZ4k&start_radio=1"

# Tiempo a recortar (segundos)
t_start = 20.0
t_end = 25.0  # tomamos 5 segundos: 20->25

# Frecuencia de muestreo objetivo (Hz)
fs_target = 44100

# Frecuencia de la portadora (Hz) - debe ser mucho mayor que el ancho de banda del mensaje.
# Elegimos 10000 Hz por defecto (si tu audio tiene mucha banda, puedes aumentarla)
fc = 10000.0

# Índice de modulación deseado (ka). El código ajusta ka para que k * max(|m|) = modulation_index
modulation_index = 1.0

# Corte de filtro pasa-bajo para demodulación (Hz)
lpf_cutoff = 5000.0

# ---------------------------
# 3) Descargar y recortar audio desde YouTube
# ---------------------------
import os
from pathlib import Path
import subprocess
from pydub import AudioSegment

workdir = "/content/am_sim"
os.makedirs(workdir, exist_ok=True)

# usar yt-dlp para bajar audio
out_mp3 = os.path.join(workdir, "audio_raw.%(ext)s")
print("Descargando audio desde YouTube (audio-only)... esto puede tardar unos segundos.")
!yt-dlp -x --audio-format mp3 -o "{out_mp3}" "{YT_URL}"  --no-playlist

# Encontrar archivo mp3 descargado
mp3_files = list(Path(workdir).glob("audio_raw.*"))
if len(mp3_files) == 0:
    # buscar otro nombre que yt-dlp pudo haber asignado
    mp3_files = list(Path(".").glob("*.mp3"))
    if len(mp3_files) == 0:
        raise RuntimeError("No se encontró archivo mp3 descargado. Revisa la URL o yt-dlp.")
mp3_path = str(mp3_files[0])
print("Archivo mp3:", mp3_path)

# Recortar 20-25s y convertir a wav mono fs_target
wav_path = os.path.join(workdir, "message_fragment.wav")
# pydub usa ms
audio = AudioSegment.from_file(mp3_path)
segment = audio[int(t_start*1000):int(t_end*1000)]
# convertir a mono y fs_target
segment = segment.set_frame_rate(fs_target).set_channels(1)
segment.export(wav_path, format="wav")
print("Fragmento guardado en:", wav_path)

# ---------------------------
# 4) Cargar el fragmento en numpy
# ---------------------------
import numpy as np
import soundfile as sf
message, fs = sf.read(wav_path)
# asegurar mono y tipo float64
if message.ndim > 1:
    message = message.mean(axis=1)
message = message.astype(np.float64)
if fs != fs_target:
    print(f"Advertencia: fs read {fs}, target {fs_target}")

t = np.arange(len(message)) / fs
print(f"Mensaje: duración {len(message)/fs:.3f} s, fs = {fs}")

# normalizar mensaje a amplitud máxima 1 (mantener relación)
m_max = np.max(np.abs(message))
if m_max == 0:
    raise RuntimeError("Mensaje es silencioso.")
message_norm = message / m_max  # ahora en [-1,1]

# ---------------------------
# 5) Preparar portadora y índice de modulación
# ---------------------------
# Queremos ka tal que ka * max(|m|) = modulation_index. Dado que message_norm max=1, ka = modulation_index
ka = modulation_index

# Amplitud de portadora A_c (elegimos 1 para simplificar)
A_c = 1.0

# portadora
carrier = A_c * np.cos(2*np.pi*fc*t)

# ---------------------------
# 6) Modulación AM (s(t) = A_c [1 + ka * m_norm(t)] cos(2π f_c t))
# ---------------------------
s_mod = A_c * (1.0 + ka * message_norm) * np.cos(2*np.pi*fc*t)

# limitar para evitar clipping (no crítico en señal continua)
s_mod = s_mod / np.max(np.abs(s_mod)) * 0.99

# Guardar audio modulada como wav para reproducción
mod_wav_path = os.path.join(workdir, "modulated.wav")
sf.write(mod_wav_path, s_mod, fs)
print("Señal modulada guardada en:", mod_wav_path)

# ---------------------------
# 7) Funciones FFT y graficado
# ---------------------------
import matplotlib.pyplot as plt

def compute_fft(x, fs):
    N = len(x)
    X = np.fft.rfft(x)
    freqs = np.fft.rfftfreq(N, d=1/fs)
    mag = np.abs(X) / N
    return freqs, mag, X

def plot_time_and_spectrum(x, fs, title_prefix="", t_zoom=None):
    N = len(x)
    t = np.arange(N)/fs
    freqs, mag, X = compute_fft(x, fs)
    plt.figure(figsize=(12,6))
    # tiempo
    plt.subplot(2,1,1)
    if t_zoom is None:
        plt.plot(t, x)
    else:
        idx0 = int(t_zoom[0]*fs)
        idx1 = int(t_zoom[1]*fs)
        plt.plot(t[idx0:idx1], x[idx0:idx1])
    plt.ylabel("Amplitud")
    plt.xlabel("Tiempo (s)")
    plt.title(f"{title_prefix} - Tiempo")
    # magnitud
    plt.subplot(2,1,2)
    plt.plot(freqs, mag)
    plt.xlim(0, fs/2)
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.title(f"{title_prefix} - Espectro (magnitud)")
    plt.tight_layout()
    plt.show()
    return freqs, mag, X

# ---------------------------
# 8) Graficar mensaje, portadora, modulada
# ---------------------------
print("Graficando mensaje original (normalizado)...")
plot_time_and_spectrum(message_norm, fs, title_prefix="Mensaje m(t)", t_zoom=(0,0.05))

print("Graficando portadora c(t)...")
plot_time_and_spectrum(carrier, fs, title_prefix=f"Portadora c(t), fc={fc} Hz", t_zoom=(0,0.005))

print("Graficando señal modulada s(t)...")
plot_time_and_spectrum(s_mod, fs, title_prefix="Señal AM s(t)", t_zoom=(0,0.01))

# ---------------------------
# 9) Reproducir fragmentos de audio (original, modulada)
# ---------------------------
from IPython.display import Audio, display

print("Reproduciendo fragmento original (mensaje)...")
display(Audio(message_norm, rate=fs))

print("Reproduciendo señal modulada (tono AM)...")
display(Audio(s_mod, rate=fs))

# ---------------------------
# 10) Demodulación coherente (multiplicar por cos y aplicar LPF ideal por FFT)
# ---------------------------
# 10.a multiplicar por portadora (coherent detection)
product = s_mod * np.cos(2*np.pi*fc*t)

# 10.b Filtrado ideal pasa-bajo implementado en dominio de la frecuencia (mask por FFT)
N = len(product)
# transformada
F = np.fft.rfft(product)
freqs = np.fft.rfftfreq(N, d=1/fs)

# diseñar máscara ideal LPF con cutoff lpf_cutoff
mask = np.zeros_like(F, dtype=np.float64)
mask[freqs <= lpf_cutoff] = 1.0

# aplicar mask
F_filtered = F * mask

# reconstruir señal demodulada (IFFT)
m_rec_time = np.fft.irfft(F_filtered, n=N)

# Factor de escala: cuando multiplicamos por cos, la componente base sale con factor A_c/2.
# Aquí A_c=1, pero además tuvimos ka y normalizaciones; ajustamos para recuperar la amplitud
# Aproximación simple: dividir por (A_c/2 * ka) para intentar recuperar m_norm
scale = 1.0 / (A_c/2.0 * ka)
m_rec_time = m_rec_time * scale

# recortar DC offset y normalizar
m_rec_time = m_rec_time - np.mean(m_rec_time)
m_rec_time = m_rec_time / (np.max(np.abs(m_rec_time)) + 1e-12)

# ---------------------------
# 11) Graficar etapas de demodulación
# ---------------------------
print("Graficando producto s(t)*cos(2πfct)")
plot_time_and_spectrum(product, fs, title_prefix="Producto (s·cos)", t_zoom=(0,0.05))

print("Graficando señal demodulada (después LPF ideal)")
plot_time_and_spectrum(m_rec_time, fs, title_prefix="Mensaje recuperado m_rec(t)", t_zoom=(0,0.05))

# ---------------------------
# 12) Reproducir mensaje recuperado
# ---------------------------
print("Reproduciendo mensaje recuperado (demodulado):")
display(Audio(m_rec_time, rate=fs))

# ---------------------------
# 13) Guardar audios finales
# ---------------------------
rec_wav_path = os.path.join(workdir, "demodulated_message.wav")
sf.write(rec_wav_path, m_rec_time, fs)
print("Mensaje demodulado guardado en:", rec_wav_path)

# ---------------------------
# 14) Mostrar resumen de archivos generados
# ---------------------------
print("\nArchivos generados en", workdir)
for f in os.listdir(workdir):
    print(" -", f)

# FIN
```
